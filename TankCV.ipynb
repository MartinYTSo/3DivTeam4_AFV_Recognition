```python
import os
import random
from ultralytics import YOLO

# Initialize the YOLO model
model = YOLO("yolo11n.pt")

# Directory containing images
image_dir = 'data'

# Get a list of all files in the directory that have 't-72' in their name
t72_images = [os.path.join(image_dir, img) for img in os.listdir(image_dir) if 't-72' in img]

# Randomly select up to 10 images from the list
random_images = random.sample(t72_images, min(100, len(t72_images)))  # Ensure you don't exceed available images

# Initialize counters for classifications
classification_counts = {'boat': 0, 'truck': 0, 'train': 0}

# Loop through the selected images and make predictions
for img_path in random_images:
    results = model.predict(source=img_path, save=True, save_dir='testimg')
    
    # Iterate through each detection result
    for result in results:
        # Get the class names for detected objects
        for cls in result.names:
            # Increment counters for relevant classes
            if cls == 'boat':
                classification_counts['boat'] += 1
            elif cls == 'truck':
                classification_counts['truck'] += 1
            elif cls == 'train':
                classification_counts['train'] += 1

# Print the classification counts
print("Classification counts:")
print(f"Boats: {classification_counts['boat']}")
print(f"Trucks: {classification_counts['truck']}")
print(f"Trains: {classification_counts['train']}")
```
